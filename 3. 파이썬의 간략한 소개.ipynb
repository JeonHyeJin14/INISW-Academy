- 대괄호 사이에 쉼표로 구분된 값들의 목록

- 인덱싱과 슬라이싱 가능

```python
squares = [1, 4, 9, 16, 25]
squares
```

```python
squares[0]  # 인덱싱은 항목을 돌려줍니다
```

```python
squares + [36, 49, 64, 81, 100]
```

```python
cubes = [1, 8, 27, 65, 125]  # 여기 뭔가 잘못 됐습니다
4 ** 3  # 4의 세제곱은 65가 아니라 64입니다!
```

```python
cubes[3] = 64  # 잘못된 값을 고칩니다
cubes
```

list.append() 메서드로 끝에 새 항목 추가 가능

```python
cubes.append(216)  # 6의 세제곱을 추가합니다
cubes.append(7 ** 3)  # 그리고 7의 세제곱도
cubes
```

= 연산자로 변수를 할당할 때 데이터 자체를 복사하는 것이 아니라 객체를 가리키는 참조를 만듦

-> 같은 객체를 여러 변수가 공유 가능

```python
rgb = ["Red", "Green", "Blue"]
rgba = rgb
id(rgb) == id(rgba)  # 같은 객체를 참조합니다
```

```python
rgba.append("Alph") #rgba와 rgb가 같은 객체 공유
rgb
```

모든 슬라이스 연산은 요청한 항목들을 포함하는 새 리스트 반환

```python
correct_rgba = rgba[:]
correct_rgba[-1] = "Alpha"
correct_rgba
```

슬라이스에 대입하는 것도 가능 -> 리스트의 길이 변경, 모든 항목 삭제 가능

```python
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
letters
```

```python
# 몇몇 값을 바꿉니다
letters[2:5] = ['C', 'D', 'E']
letters
```

```python
# 이제 그 것들을 지웁니다
letters[2:5] = [] #빈 리스트로 설정
letters
```

```python
# 모든 요소들을 빈 리스트로 치환해서 리스트를 비웁니다
letters[:] = []
letters
```

```python
# 내장함수 len() 적용
letters = ['a', 'b', 'c', 'd']
len(letters)
```

# 프로그래밍으로의 첫 걸음

피보나치 수열

```python
# 피보나치 수열:
# 두 요소의 합이 다음을 정의합니다
a, b = 0, 1
while a < 10:
    print(a)
    a, b = b, a+b
```

#### 다중 대입
: 변수 a와 b에 동시에 0과 1이 대입됨

#### while루프
: 조건이 참인동안 실행

#### 루프의 body는 들여쓰기
: 들여쓰기는 파이썬에 문장을 덩어리로 묶는 방법

#### print()
: 주어진 인자들의 값을 인쇄하는 함수

```python
i = 256*256
print('The value of i is', i)
```

키워드 인자 end는 출력 끝에 포함되는 개행문자를 제거하거나 출력을 다른 문자열로 끝나게 하고 싶을 때 사용

```python
a, b = 0, 1
while a < 1000:
    print(a, end=',')
    a, b = b, a+b
```
